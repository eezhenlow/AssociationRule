```python
# Bakery Sales

import pandas as pd

# Load the uploaded CSV file
file_path = 'Bakery_Sales.csv'
sales_df = pd.read_csv(file_path)
                       
# Transform the data to create a one-hot encoded format for items by transaction
pivot_df_bakery_transformed = sales_df.pivot_table(index='Transaction', columns='Item', aggfunc='size', fill_value=0).reset_index(drop=True)

# Convert values to 1 if they are >=1, otherwise 0
pivot_df_bakery_transformed = pivot_df_bakery_transformed.applymap(lambda x: 1 if x >= 1 else 0)

# Set the target row count
target_row_count = 3800  # Change this to your desired row count

# Check if the DataFrame needs to be reduced
if len(pivot_df_bakery_transformed) > target_row_count:
    # Randomly sample the DataFrame to the target row count
    pivot_df_bakery_transformed = pivot_df_bakery_transformed.sample(n=target_row_count, random_state=1).reset_index(drop=True)

# Save the transformed dataframe to a new CSV file
output_path = 'Bakery_Sales_Transformed.csv'
pivot_df_bakery_transformed.to_csv(output_path, index=False)

# Display the transformed dataframe to the user
pivot_df_bakery_transformed.head()
```

```python
# Online Retail Sales

import pandas as pd

# Load the uploaded CSV file
file_path = 'Online_Retail_Sales.xlsx'
sales_df = pd.read_excel(file_path)

# Filter out rows where the specified column contains 'C'
df_filtered = sales_df[~sales_df['InvoiceNo'].str.contains('C', na=False)]

df_filtered['StockCode'] = df_filtered['StockCode'].astype(str)

# Transform the data to create a one-hot encoded format for items by transaction
pivot_df_online_retail_transformed = df_filtered.pivot_table(index='InvoiceNo', columns='StockCode', aggfunc='size', fill_value=0).reset_index(drop=True)

# Convert values to 1 if they are >=1, otherwise 0
pivot_df_online_retail_transformed = pivot_df_online_retail_transformed.applymap(lambda x: 1 if x >= 1 else 0)

# Set the target row count
target_row_count = 3800  # Change this to your desired row count

# Check if the DataFrame needs to be reduced
if len(pivot_df_online_retail_transformed) > target_row_count:
    # Randomly sample the DataFrame to the target row count
    pivot_df_online_retail_transformed = pivot_df_online_retail_transformed.sample(n=target_row_count, random_state=1).reset_index(drop=True)

# Drop columns where all values are 0
pivot_df_online_retail_transformed = pivot_df_online_retail_transformed.loc[:, (pivot_df_online_retail_transformed != 0).any(axis=0)]

# User-specified target column count
target_column_count = 900

# Calculate the occurrence (count of non-zero values) for each column
occurrence_counts = (pivot_df_online_retail_transformed != 0).sum(axis=0)

# Sort columns by occurrence count in ascending order
columns_to_keep = occurrence_counts.nlargest(target_column_count).index

# Keep only the target number of columns with the highest occurrences
pivot_df_online_retail_transformed = pivot_df_online_retail_transformed[columns_to_keep]

# Save the transformed dataframe to a new CSV file
output_path = 'Online_Retail_Sales_Transformed.csv'
pivot_df_online_retail_transformed.to_csv(output_path, index=False)

# Display the transformed dataframe to the user
pivot_df_online_retail_transformed.head()
```

```python
import time

# Load the re-uploaded transformed dataset
file_path = 'Groceries_Sales_Transformed.csv'
bakery_sales_df = pd.read_csv(file_path)

# Import necessary libraries for Apriori analysis
from mlxtend.frequent_patterns import apriori, association_rules

# Perform Apriori algorithm to find frequent itemsets
# Set a minimum support threshold
min_support = 0.01
min_threshold = 0.2

# Apply apriori algorithm to get frequent itemsets
start_time = time.time()
frequent_itemsets = apriori(bakery_sales_df, min_support=min_support, use_colnames=True)

# Generate association rules from the frequent itemsets
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=min_threshold)

end_time = time.time() - start_time 

# Add a column 'k' representing the size of each itemset
frequent_itemsets['k'] = frequent_itemsets['itemsets'].apply(len)

print(frequent_itemsets['k'].mean())
print(end_time)

# Display the itemsets with their corresponding k value
print(frequent_itemsets)

frequent_itemsets.head(), rules.head()
```

```python
import itertools
import pandas as pd
import time

# Load the re-uploaded transformed dataset
file_path = 'Bakery_Sales_Transformed_Test.csv'
df = pd.read_csv(file_path)

# Set support and confidence thresholds
min_support = 0.001
min_confidence = 0.2

# Get all possible item combinations
start_time = time.time()
items = df.columns
all_itemsets = []
for r in range(1, len(items) + 1):
    all_itemsets.extend(itertools.combinations(items, r))

# Calculate support for each itemset
itemset_support = {}
for itemset in all_itemsets:
    support = (df[list(itemset)].all(axis=1).sum()) / len(df)
    if support >= min_support:
        itemset_support[itemset] = support

# Generate association rules
rules = []
for itemset, support in itemset_support.items():
    if len(itemset) > 1:
        # Generate all possible antecedent and consequent pairs
        for i in range(1, len(itemset)):
            for antecedent in itertools.combinations(itemset, i):
                consequent = tuple(sorted(set(itemset) - set(antecedent)))
                # Calculate confidence
                antecedent_support = itemset_support.get(antecedent, 0)
                if antecedent_support > 0:
                    confidence = support / antecedent_support
                    if confidence >= min_confidence:
                        rules.append({
                            'antecedent': antecedent,
                            'consequent': consequent,
                            'support': support,
                            'confidence': confidence
                        })

end_time = time.time() - start_time
# Convert to DataFrame for easier viewing
rules_df = pd.DataFrame(rules)
print("Association Rules:")
print(rules_df)
print(end_time)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for plotting
dataset_names = ["Bakery", "Grocery", "Online Retail", "E-Commerce", "Market Basket"]
apriori_times = [0.024, 1.31, 53, 43, 101]  # In seconds
brute_force_times = [7.14e49, 5.11e93, np.inf, np.inf, np.inf]  # Infinite values for large computations

# Set up the plot
fig, ax = plt.subplots(figsize=(10, 6))

# Plot Apriori Algorithm times
ax.plot(dataset_names, apriori_times, label="Apriori Algorithm", marker='o', color='blue', linewidth=2, markersize=6)

# Convert brute-force times for plotting (replace 'inf' with a very high value for visualization)
brute_force_display = [val if val != np.inf else 1e100 for val in brute_force_times]
ax.plot(dataset_names, brute_force_display, label="Brute-Force Algorithm", marker='x', color='red', linewidth=2, markersize=6)

# Set y-axis to logarithmic scale
ax.set_yscale('log')
ax.set_ylim([1e-2, 1e105])  # Adjusting limits to make Apriori times visible

# Add labels and title
ax.set_xlabel("Dataset Name")
ax.set_ylabel("Time (Seconds, Log Scale)")
ax.set_title("Performance Comparison: Apriori vs. Brute-Force Algorithms")
ax.legend()

plt.show()
```
